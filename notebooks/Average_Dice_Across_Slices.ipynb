```python
## Load Imports ##
import nibabel as nib
import numpy as np
from PIL import Image

import os
import sys
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)
from metrics import dice_coef, rec_score, prec_score
```

```python
import matplotlib.pyplot as plt
def thresholdMask(mask, threshold=0.56):
    mask[mask >= threshold] = 1.0
    mask[mask < threshold] = 0.0
    return mask

def applyMask(mask, normal):
    return(np.multiply(mask[:, :, 0], normal))

def show_slices(slices):
    """ Function to display row of image slices """
    fig, axes = plt.subplots(1, len(slices))
    for i, slice in enumerate(slices):
      axes[i].imshow(slice.T, cmap="gray", origin="lower")

def show_slice(slice):
    plt.imshow(slice.T, cmap="gray", origin="lower")
    
def normalizeImg(x):
    # Normalize x
    max_val = np.max(x)
    min_val = np.min(x)
    norm_x = (x-min_val)/(max_val-min_val + 1e-7)
    return norm_x
```

```python
mask_dir = '../slice_data_side_test_pred/'
pred_dir = '../slice_data_side_test/'
thresh = 0.60 #52
side_dc = []
for _,_,files in os.walk(mask_dir):
    for file in files:
        # Construct normal_file path
        pred_file = (pred_dir + file)
        pred_file = pred_file.replace('mask_pred_', 'mask_')
        # Construct file path
        mask_file = (mask_dir + file)
        #print('-------------------')
        #print(pred_file)
        #print(mask_file)
        #print('-------------------')
        mask_data = ((nib.load(mask_file)).get_data())[:,:,0]
        pred_data = (nib.load(pred_file)).get_data()
        #show_slice(mask_data)
        #plt.show()
        #show_slice(pred_data)
        #plt.show()
        dc = dice_coef(pred_data, mask_data, threshold=thresh)
        #print(dc)
        if dc != -1:
            side_dc.append(dc)

np.mean(side_dc)
```
